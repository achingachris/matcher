```python
import pandas as pd

MENTOR_DATA = './data/mentors.csv'
MENTEE_DATA = './data/mentee.csv'
MATCHES_FILE = './data/matches.csv'

def read_data(filename):
    """
    Read data from a CSV file efficiently.
    """
    data = pd.read_csv(filename)
    data.columns = data.columns.str.strip().str.lower()
    data['id'] = data.index
    data['id'] = data['id'].apply(lambda x: f'{filename}-{x}')
    print(f"{filename} > {data.shape}")
    return data
```

```python
read_data(MENTOR_DATA)
read_data(MENTEE_DATA)
# read_data(MATCHES_FILE)
```

```python
def match_mentors_mentees(mentors, mentees):
    """
    Match mentors with mentees based on predefined fields.
    """
    matches = []

    for m_index, mentee in mentees.iterrows():
        for t_index, mentor in mentors.iterrows():
            skills_match = any(skill in mentee['skills'] for skill in mentor['skills'])
            demographic_match = mentee['age'] - 5 <= mentor['age'] <= mentee['age'] + 5

            if skills_match and demographic_match:
                matches.append({
                    'mentee_id': mentee['id'],
                    'mentor_id': mentor['id'],
                    'skills_match': skills_match,
                    'demographic_match': demographic_match
                })

    return pd.DataFrame(matches)
```

```python
def main():
    print("Reading in two dataframes to match ...")
    mentors = read_data(MENTOR_DATA)
    mentees = read_data(MENTEE_DATA)

    print("Matching mentors and mentees ...")
    matches = match_mentors_mentees(mentors, mentees)

    print(f"Saving matches to {MATCHES_FILE}")
    matches.to_csv(MATCHES_FILE, index=False)
    print("Matching completed.")

if __name__ == "__main__":
    main()
```
